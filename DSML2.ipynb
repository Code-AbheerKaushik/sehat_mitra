```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
```

```python
# Step 1: Load dataset
```

```python
df = pd.read_csv("/Users/heisenberg/Desktop/Coding/College ka kaam /DSML/students.csv")
print("Original Dataset:\n")
df
```

```python
# Step 2: Handle missing numerical data (Age, Score) using SimpleImputer (mean)
num_imputer = SimpleImputer(strategy="median")
df[['Age', 'Score']] = num_imputer.fit_transform(df[['Age', 'Score']])
```

```python
# Step 3: Handle missing categorical data (Gender, City) using Mode
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['City'] = df['City'].fillna(df['City'].mode()[0])
```

```python
# Step 4: Label Encoding for categorical columns
le_gender = LabelEncoder()
df['Gender_label'] = le_gender.fit_transform(df['Gender'].astype(str))
```

```python
le_city = LabelEncoder()
df['City_label'] = le_city.fit_transform(df['City'].astype(str))
```

```python
# Step 5: One-Hot Encoding (Dummy Variables) for categorical columns
df_dummy = pd.get_dummies(df[['Gender', 'City']], prefix=['Gender', 'City'])
```

```python
# Step 6: Combine original columns, label encoded, and dummy variables
df_final = pd.concat([df, df_dummy], axis=1)
```

```python
# Step 7: Display final dataset
print("Final Dataset with Original, Label Encoded, and Dummy Variables:")
df_final
```
